```python
import folium
import pandas as pd
from folium.features import DivIcon
```

```python
df = pd.read_excel('teamfs.xlsx',header=None,names=["team_name","latitude","longtitude","city",
                                                    "year_of_creation","year_of_defunct","standings"],skiprows=1)
df
```

```python
df=df.astype({"team_name":str,"latitude":float,"longtitude":float,"city":str,"year_of_creation":str,"year_of_defunct":str,"standings":str})
df
```

```python
m = folium.Map(location=[37.090240,-90.712891],zoom_start=4)
```

```python
t=[]
for i in range(0,len(df.index),1):
    t.append(df.at[i,"standings"])
t
    
    
```

```python
def koncowki(x):
    k=[]
    for i in range(0,len(x),1):
        if(x[i]=="1"):
            k.append("st")
        elif(x[i]=="2"):
            k.append("nd")
        elif(x[i]=="3"):
            k.append("rd")
        else:
            k.append("th")
    return k
        
```

```python
x=koncowki(t)
x
```

```python
def kolory(x):
    if (x=="st"):
        return "orange"
    else:
        return "blue"
```

```python
def ikona(x):
    if (x=='st'):
        return "trophy"
    else:
        return "map-marker"
```

```python


for i in range(0,11,1):
    folium.Marker(location=[df.at[i,"latitude"],df.at[i,"longtitude"]],popup="Team: " 
                  + df.at[i,"team_name"]+",<br>"+"Rok year_of_creation: " + 
              df.at[i,"year_of_creation"]+",<br>"+"year_of_defunct: " + df.at[i,"year_of_defunct"]+",<br>"+
                  "standings: " + df.at[i,"standings"]+x[i], 
                  tooltip="City: " + df.at[i,"city"]+",<br>"+"Click to get more information",
                  icon=folium.Icon(icon=ikona(x[i]),color=kolory(x[i]),icon_color='#FFFFFF',prefix='fa')).add_to(m) 

div = folium.DivIcon(html=(
    '<svg height="105" width="230">'
    '<text x="10" y="30" font-size=25px font-style="italic" fill="black">Map of cities</text>'
    '<text x="10" y="55" font-size=25px font-style="italic" fill="black">with teams in</text>'
    '<text x="10" y="80" font-size=25px font-style="italic" fill="black">season 1946-1947</text>'
    '</svg>'
))
folium.Marker((36, -70), icon=div).add_to(m)
m.save("map1946.html")
```
